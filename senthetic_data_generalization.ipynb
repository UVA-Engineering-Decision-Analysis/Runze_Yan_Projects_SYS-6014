```python
def wgn(x, snr):
    snr = 10 ** (snr / 10.0)
    xpower = np.sum(x ** 2) / len(x)
    npower = xpower / snr
    
    return np.random.randn(len(x)) * np.sqrt(npower)
```

```python
plt.style.use('fivethirtyeight')
plt.figure('Sythetic Data', figsize=(30,6))
X = np.linspace(0,300,300)
Y = 150 * np.sin(np.pi * X / 3) + 180 * np.cos(np.pi * X / 4)
SNR = 6
N = wgn(X, SNR)
Y =Y + N

plt.plot(X,Y)
```
